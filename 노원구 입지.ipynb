```python
# 노원구에 있는 커피 전문점
df = df[(df['업태구분명']=='커피숍') | (df['업태구분명']=='다방')| (df['업태구분명']=='까페')]
df = df[(df['영업상태명']=='영업/정상')]
df_coffee = df.loc[df.소재지전체주소.str.contains("노원구", na=False)]
```

```python
df_coffee = df_coffee[~df_coffee['사업장명'].str.contains("스타벅스")|~df_coffee['사업장명'].str.contains("스타벅스")|~df_coffee['사업장명'].str.contains("스타벅스")|~df_coffee['사업장명'].str.contains("스타벅스")|~df_coffee['사업장명'].str.contains("투썸플레이스")|~df_coffee['사업장명'].str.contains("폴바셋")|~df_coffee['사업장명'].str.contains("파스쿠찌")|~df_coffee['사업장명'].str.contains("엔제리너스")|~df_coffee['사업장명'].str.contains("할리스")]
```

```python
from pyproj import Proj, transform
import pandas as pd

# Projection 정의
# 중부원점(Bessel): 서울 등 중부지역 EPSG:2097
proj_1 = Proj(init='epsg:2097')

# WGS84 경위도: GPS가 사용하는 좌표계 EPSG:4326
proj_2 = Proj(init='epsg:4326')

DataFrame = df_coffee.copy()

x_list = []
y_list = []

for idx, row in DataFrame.iterrows():
    x, y = row['좌표정보(x)'], row['좌표정보(y)']
    x_, y_ = transform(proj_1, proj_2, x, y)
    x_list.append(x_)
    y_list.append(y_)
    
df_coffee['경도'] = x_list
df_coffee['위도'] = y_list
```

## 2. 노원구 공원을 지도에 표시

```python
import pandas as pd
df_gar = pd.read_excel('C:/Users/jwkan/OneDrive/바탕 화면/융복합 빅데이터/노원구 선정 공원.xlsx')
df_gar.set_index('공원명', inplace = True)
df_gar
```

```python
import pandas as pd
import folium

# 서울 지도 만들기
seoul_map = folium.Map(location=[37.55,126.98], tiles='Stamen Terrain', 
                        zoom_start=12)

# 공원 위치정보를 CircleMarker로 표시
for name, lat, lng in zip(df_coffee.사업장명, df_coffee.위도, df_coffee.경도):
    folium.CircleMarker([lat, lng],
                        radius=3,         # 원의 반지름
                        color='green',         # 원의 둘레 색상
                        fill=True,
                        fill_color='green',    # 원을 채우는 색
                        fill_opacity=1, # 투명도    
                        popup=name
    ).add_to(seoul_map)


for name, lat, lng in zip(df_gar.index, df_gar.위도, df_gar.경도):
    folium.Marker([lat, lng], icon=folium.Icon('blue', icon='star'),
                  popup=name
                 ).add_to(seoul_map)
    
for name, lat, lng in zip(df_gar.index, df_gar.위도, df_gar.경도):
    folium.CircleMarker([lat, lng],
                        radius=30,         # 원의 반지름
                        color='#3186cc',         # 원의 둘레 색상
                        fill=True,
                        fill_color='#3186cc',    # 원을 채우는 색   
                        popup=name
    ).add_to(seoul_map)

display(seoul_map)
        

# 지도를 HTML 파일로 저장하기
seoul_map.save('output/seoul_colleges2.html')
```
