### 열별로 minmax scaling 진행 한 후 항목별로 가중치를 달리하여 합을 구함.
### 총합 = 커피소비인구수 X 0.1 + 테이크 아웃 전문점 수 X 0.2 + 20대 유동인구 X 0.3 + 대학생 수 X 0.4
